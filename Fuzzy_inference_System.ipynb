```python
#Libraries
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
#Functions
#sigmoid
def sigmoidMF(x,a,c):
  return (1/(1+(np.e**(-a*(x-c)))))

#Gauss
def gaussMF(x,a,c):
  return np.e**(-(x-c)**2/(2*a**2))

#bell
def bellMF(x,a,b,c):
  return 1/(1+(np.absolute((x-c)/a))**(2*b))

#triangular
def triangularMF(x,a,b,c):
  return np.maximum(np.minimum((x-a)/(b-a),(c-x)/(c-b)),0)

#trapzoidal
def trapMF(x,a,b,c,d):
  #return np.maximum(np.minimum((x-a/b-a),1,(d-x/d-c)),0)
  #return np.maximum(np.minimum(np.minimum((x-a/b-a),1),np.minimum(1,(d-x/d-c))),0)
  
  y = np.ones(len(x))

  idx = np.nonzero(x <= b)
  y[idx] = triangularMF(x[idx],a,b,b)

  idx = np.nonzero(x >= c)
  y[idx] = triangularMF(x[idx],c,c,d)

  idx = np.nonzero(x < a)
  y[idx] = np.zeros(len(idx))

  idx = np.nonzero(x > d)
  y[idx] = np.zeros(len(idx))
  
  return y
```

```python
#Universe(s)
x = np.linspace(0,359,100) #flag/generator
y = np.linspace(0,100,100) #speed
z = np.linspace(0,5,100)   #direction

#Membership Functions
#Flag position
EN = trapMF(x,0,0,10,80)
N = trapMF(x,10,80,100,170)
W = trapMF(x,100,170,190,260)
S = trapMF(x,190,260,280,350)
E = trapMF(x,280,350,359,359)

#Generator position
ENg = trapMF(x,0,0,10,80)
Ng = trapMF(x,10,80,100,170)
Wg = trapMF(x,100,170,190,260)
Sg = trapMF(x,190,260,280,350)
Eg = trapMF(x,280,350,359,359)

#Speed
L = trapMF(y,0,0,10,35)
M = trapMF(y,15,40,60,85)
H = trapMF(y,65,90,100,100) 

#Direction
CW = triangularMF(z,0,0,2.4)
DM = triangularMF(z,2.3,2.5,2.7)
CCW = triangularMF(z,2.6,5,5)

#Plot
plt.figure(figsize=(8, 13), dpi= 80, facecolor='w', edgecolor='k')

#Figure 1
plt.subplot(4,1,1)
plt.plot(x,EN)
plt.plot(x,N)
plt.plot(x,W)
plt.plot(x,S)
plt.plot(x,E)
plt.title('Flag input')
plt.gca().legend(('EN','N','W','S','E'))
plt.grid(True)

#Figure 2
plt.subplot(4,1,2)
plt.plot(x,ENg)
plt.plot(x,Ng)
plt.plot(x,Wg)
plt.plot(x,Sg)
plt.plot(x,Eg)
plt.title('Generator position')
plt.gca().legend(('EN','N','W','S','E'))
plt.grid(True)

#Figure 3
plt.subplot(4,1,3)
plt.plot(y,L,'r')
plt.plot(y,M,'g')
plt.plot(y,H,'b')
plt.title('Speed')
plt.gca().legend(('Low','Med','High'))
plt.grid(True)

#Figure 4
plt.subplot(4,1,4)
plt.plot(z,CW,'r')
plt.plot(z,DM,'g')
plt.plot(z,CCW,'b')
plt.title('Direction')
plt.gca().legend(('Clock Wise','Dont Move','Counter Clock Wise'))
plt.grid(True)

plt.show()
```

```python
FlagIn = ['EN','N','W','S','E']
GenPos = ['EN','N','W','S','E']

Output = np.empty((5,5),dtype='object')
indx = -1
for x in FlagIn:
  indx = indx + 1
  indy = -1
  for y in GenPos:
    indy = indy +1
    if x == y:
      Output[indx][indy] = L,DM
      
    if x == 'N' and y == 'S' or x == 'S' and y == 'N':
      Output[indx][indy] = H,CCW
      
    if x == 'E' and y == 'W' or x == 'W' and y == 'E':
      Output[indx][indy] = H,CW
      
    if x == 'EN' and y == 'W' or x == 'W' and y == 'EN':
      Output[indx][indy] = H,CCW
      
    if x == 'EN' and y == 'N'or x == 'N' and y == 'EN':
      Output[indx][indy] = H,CW
      
    if x == 'S' and y == 'EN' or x == 'EN' and y == 'S':
      Output[indx][indy] = M,CCW
      
    if x == 'W' and y == 'N' or x == 'N' and y == 'W':
      Output[indx][indy] = M,CCW
      
    if x == 'EN' and y == 'E' or x == 'E' and y == 'EN':
      Output[indx][indy] = L,CCW
      
    if x == 'N' and y == 'E' or x == 'E' and y == 'N':
      Output[indx][indy] = M,CCW
      
    if x == 'S' and y == 'W' or x == 'W' and y == 'S':
      Output[indx][indy] = M,CW
      
    if x == 'E' and y == 'S' or x == 'S' and y == 'E':
      Output[indx][indy] = M,CCW

output = np.empty((5,5),dtype='object')
indx = -1
for x in FlagIn:
  indx = indx + 1
  indy = -1
  for y in GenPos:
    indy = indy +1
    if x == y:
      output[indx][indy] = 'L','DM'
      
    if x == 'N' and y == 'S' or x == 'S' and y == 'N':
      output[indx][indy] = 'H','CCW'
      
    if x == 'E' and y == 'W' or x == 'W' and y == 'E':
      output[indx][indy] = 'H','CW'
      
    if x == 'EN' and y == 'W' or x == 'W' and y == 'EN':
      output[indx][indy] = 'H','CCW'
      
    if x == 'EN' and y == 'N'or x == 'N' and y == 'EN':
      output[indx][indy] = 'H','CW'
      
    if x == 'S' and y == 'EN' or x == 'EN' and y == 'S':
      output[indx][indy] = 'M','CCW'
      
    if x == 'W' and y == 'N' or x == 'N' and y == 'W':
      output[indx][indy] = 'M','CCW'
      
    if x == 'EN' and y == 'E' or x == 'E' and y == 'EN':
      output[indx][indy] = 'L','CCW'
      
    if x == 'N' and y == 'E' or x == 'E' and y == 'N':
      output[indx][indy] = 'M','CCW'
      
    if x == 'S' and y == 'W' or x == 'W' and y == 'S':
      output[indx][indy] = 'M','CW'
      
    if x == 'E' and y == 'S' or x == 'S' and y == 'E':
      output[indx][indy] = 'M','CCW'

print(output)

```

# Fuzz and defuzz

```python
Ze = np.empty((len(X),len(X)),dtype='object')

for ev1,t in enumerate(X):
  for ev2,s in enumerate(X):
    indx = -1
    ind = 0
    outL = []
    for x in FlagIn:
      indx = indx + 1
      indy = -1
      c = []
      for y in GenPos:
        indy = indy +1
        indz = -1

        for z in Output:
          indz = indz + 1
          ind = ind + 1

          if indz == 0:
            a1 = flagin[indx][ev1]

          if indz == 1:
            a2 = genpos[indy][ev2]

          if indz == 2:
            c1 = np.minimum(a1,a2)

          if indz == 3:
            c2 = np.minimum(a1,a2)
          else:
            c1 = 0
            c2 = 0

          Outline = np.maximum(c1,c2)
          a = np.full(len(Z),Outline)
          outL.append(np.minimum(a,Output[indx,indy][0]))
          
        FinalCutLine = np.amax(outL,axis = 0)
        sum_x = np.nansum(FinalCutLine*Y)
        sum_y = np.nansum(FinalCutLine) + 1
        CentroidX = (sum_x/sum_y)
        Ze[ev1,ev2] = CentroidX
        
        
```

```python
Z2 = np.empty((len(X),len(X)),dtype='object')

for ev1,t in enumerate(X):
  for ev2,s in enumerate(X):
    indx = -1
    ind = 0
    outL = []
    for x in FlagIn:
      indx = indx + 1
      indy = -1
      c = []
      for y in GenPos:
        indy = indy +1
        indz = -1

        for z in Output:
          indz = indz + 1
          ind = ind + 1

          if indz == 0:
            a1 = flagin[indx][ev1]

          if indz == 1:
            a2 = genpos[indy][ev2]

          if indz == 2:
            c1 = np.minimum(a1,a2)

          if indz == 3:
            c2 = np.minimum(a1,a2)
          else:
            c1 = 0
            c2 = 0

          Outline = np.maximum(c1,c2)
          a = np.full(len(Y),Outline)
          outL.append(np.minimum(a,Output[indx,indy][1]))
          
        FinalCutLine = np.amax(outL,axis = 0)
        length = FinalCutLine.shape[0]
        sum_x = np.nansum(FinalCutLine*Z)
        sum_y = np.nansum(FinalCutLine) + 1
        CentroidX = (sum_x/sum_y)
        #print(CentroidX)
        Z2[ev1,ev2] = CentroidX
```
